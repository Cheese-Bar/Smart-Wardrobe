```python
import pandas as pd
import numpy as np

data = pd.read_excel('./data/weather_cloth_data_30.xlsx',index_col=0)
data.head()
```

```python
data=data[['1、外套(最低温度)','1、外套(最高温度)','1、服装款式与对应温度范围的调查—短袖(最低温度)','1、短袖(最高温度)','1、棉袄(最低温度)','1、棉袄(最高温度)','1、毛衣(最低温度)','1、毛衣(最高温度)','1、长袖(最低温度)','1、长袖(最高温度)']]
data
```

```python
data.columns=['outwear-low','outwear-high','Tshirt-low','Tshirt-high','coat-low','coat-high','pullover-low','pullover-high','shirt-low','shirt-high']
data.head()
```

```python
outwear = data[['outwear-low','outwear-high']].values
# print(outwear)
shortsleeve=data[['Tshirt-low','Tshirt-high']].values
coat = data[['coat-low','coat-high']].values
pullover = data[['pullover-low','pullover-high']].values
longsleeve = data[['shirt-low','shirt-high']].values
```

```python
#short 0, long 1, outwear 2, pullover 3, coat 4
import numpy as np

outwear2 = np.array([])
for i in outwear:
    xp = [0,5]
    x = [0,1,2,3,4,5]
    result = np.interp(x,xp,i)
    outwear2=np.append(outwear2,result)
    
# print(outwear2.shape)

shortsleeve2 = np.array([])
for i in shortsleeve:
    xp = [0,5]
    x = [0,1,2,3,4,5]
    result = np.interp(x,xp,i)
    shortsleeve2=np.append(shortsleeve2,result)

coat2 = np.array([])
for i in coat:
    xp = [0,5]
    x = [0,1,2,3,4,5]
    result = np.interp(x,xp,i)
    coat2=np.append(coat2,result)

longsleeve2 = np.array([])
for i in longsleeve:
    xp = [0,5]
    x = [0,1,2,3,4,5]
    result = np.interp(x,xp,i)
    longsleeve2=np.append(longsleeve2,result)

pullover2 = np.array([])
for i in pullover:
    xp = [0,5]
    x = [0,1,2,3,4,5]
    result = np.interp(x,xp,i)
    pullover2=np.append(pullover2,result)

# print(pullover2)
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt 
from scipy import stats
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split 

X = np.array([])
y = np.array([])
X = np.append(X,shortsleeve2)
y = np.append(y,np.zeros_like(shortsleeve2,dtype=np.int32))
X = np.append(X,longsleeve2)
y = np.append(y,np.ones_like(longsleeve2,dtype=np.int32))
X = np.append(X,outwear2)
y = np.append(y,np.full_like(outwear2,2,dtype=np.int32))
X = np.append(X,pullover2)
y = np.append(y,np.full_like(pullover2,3,dtype=np.int32))
X = np.append(X,coat2)
y = np.append(y,np.full_like(coat2,4,dtype=np.int32))
print(y)
X1 = X.reshape(-1, 1)
y1 = y.reshape(-1,1)
```

```python

X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.3, shuffle=True)
clf = SVC(decision_function_shape='ovo',verbose=True)
clf.fit(X_train,y_train)
```

```python

result = clf.score(X_test,y_test)
print(result)
```

```python
temp = np.linspace(-20,40)

for i in temp:
    print(i,clf.predict([[i]]))

```
